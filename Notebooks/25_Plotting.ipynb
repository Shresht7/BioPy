# Plotting Sequence Length and GC Content

```python
from Bio import SeqIO
```

```python
sequences = []
for record in SeqIO.parse("ls_orchid.fasta", "fasta"):
    sequences.append(record.seq)
```

```python
sequences
```

## Histogram of sequence lengths

```python
import matplotlib.pyplot as plt

sizes = [len(rec) for rec in sequences]

plt.hist(sizes, bins=20)
plt.title(f"{len(sizes)} orchid sequences\nLengths {min(sizes)} to {max(sizes)}")
plt.xlabel("Sequence Length (bp)")
plt.ylabel("Count")
plt.show()
```

Notice that a lot of these orchid sequences are about 740bp long!

## Plot of sequence GC%

```python
from Bio.SeqUtils import gc_fraction

gc_percentages = sorted((gc_fraction(rec) * 100) for rec in sequences)
```

```python
plt.plot(gc_percentages)
plt.title(f"{len(gc_percentages)} orchid sequences\nGC% {min(gc_percentages)} to {max(gc_percentages)}")
plt.xlabel("Genes")
plt.ylabel("GC%")
plt.show()
```
